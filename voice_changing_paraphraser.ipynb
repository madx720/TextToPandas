```python
merged_df = pd.concat([df,df1,df2])
```

```python
merged_df.reset_index(drop=True,inplace=True)
```

```python
merged_df.drop_duplicates(subset=['question'],inplace=True)
```

```python
merged_df.dropna(subset=['question'],inplace=True)
```

```python
from ast import literal_eval

for index,row in merged_df.iterrows():
    # print(index)
    tables = literal_eval(row['''sql'''])['from']['table_units']
    input = row['question'].lower()
    if 'sql' in tables[0]:
        tables = tables[0][1]['from']['table_units']
    # print(tables)
    for i,j in tables:
        table_name = list(literal_eval(row['tables'])[j].keys())[0]

        # print(table_name)
        input = input+' | '+table_name+' : '
        cols = list(literal_eval(row['tables'])[j].values())[0]
        # print(cols)
        input = input+cols[0]
        for k in cols[1:]:
            input = input+', '+k
        # print(input)
    merged_df.loc[index,'input'] = input
```
